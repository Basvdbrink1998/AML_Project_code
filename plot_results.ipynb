```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import seaborn as sns

sns.set_context("poster")
plt.style.use('seaborn-poster')

font = {'size'   : 50, 'weight' : 'normal'}

plt.rc('font', **font)

plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.bottom'] = True
plt.rc('axes',edgecolor='#808080')
```

```python
results = {
            "InceptionV3": 0.58016,
            "Voting system" : 0.578,
            "Densenet" : 0.5502,
            "MNV2" : 0.501
```

```python
plt.rcParams.update({'font.size': 50})
```

```python
plt.bar(range(len(results)), results.values(), tick_label=list(results.keys()), color=['powderblue', 'moccasin', 'mistyrose', '#adebad'])
plt.yticks(fontsize=22, color='#808080')
plt.xticks(fontsize=22, color='#808080')
plt.ylabel("Accuracy", fontsize=50, color='#808080')
plt.savefig("final_results.png")
```
